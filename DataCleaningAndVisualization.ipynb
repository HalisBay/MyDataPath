```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
y = np.arange(-10, 11)
```

```python
plt.figure(figsize=(12, 6))

plt.title('somethings')

plt.plot(y, y ** 2)
plt.plot(y, -1 * (y ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1)
plt.plot(y, y ** 2)
plt.plot([0, 0, 0], [-100, 0, 100])
plt.legend(['Y^2', 'Vertical Line'])
plt.xlabel('Y')
plt.ylabel('Y Squared')
plt.subplot(1, 2, 2)
plt.plot(y, -1 * (y ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
x = np.arange(-10, 11)
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=2,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b o', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()

fig
```

```python

fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid', color = 'red')
axes.plot(x, x + 1, linestyle='dashed', color = 'gray')
axes.plot(x, x + 2, linestyle='dashdot',color = 'blue')
axes.plot(x, x + 3, linestyle='dotted', color = 'green');

axes.set_title("My Nice Plot")
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-.b', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("My Nice Plot")

axes.legend()
```
